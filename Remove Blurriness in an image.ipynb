```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

img = cv2.imread("data_5/2.png")
b,g,r = cv2.split(img)           # get b,g,r
rgb_img = cv2.merge([r,g,b])     # switch it to rgb

# Denoising

dst = cv2.fastNlMeansDenoisingColored(img,None,10,10,7,21) 
b,g,r = cv2.split(dst)           # get b,g,r
rgb_dst = cv2.merge([r,g,b])     # switch it to rgb


cv2.imshow('denoising black and white', rgb_dst)
cv2.waitKey(0)
cv2.imwrite('data_5/small_result.png',rgb_dst)
```

```python
import cv2 as cv

#input

img = cv.imread('data_5/1.png',0)

#gaussian Blur

img = cv.GaussianBlur(img, (15,15),0)

#adaptive threshold

th3 = cv.adaptiveThreshold(img,255,cv.ADAPTIVE_THRESH_GAUSSIAN_C,\

cv.THRESH_BINARY,11,2)

cv2.imshow('Noise Filtered Image', th3)
cv2.waitKey(0)
cv.imwrite('data_5/result.png',th3)

```
